# **Now removing outliers using IQR technique**

**finding out lower and upper boind for the values in dataset**

```python
q1 = df['sepal_width'].quantile(0.25)
q3 = df['sepal_width'].quantile(0.75)
iqr = q3-q1
LowerBound = q1 - 1.5*iqr
UpperBound = q3 + 1.5*iqr
print(LowerBound, UpperBound)
```

**removing the outliers**

```python
df = df[(df['sepal_width'] >= LowerBound) & (df['sepal_width'] <= UpperBound)]
df.shape
```

# **Finding out the optimum no. of clusters for K-means using Elbow method**

**Elbow method is used to find the no. of clusters**

```python
x = df.iloc[: , [0, 1, 2, 3]].values
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',
                   max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

**Now plotting the results**

```python
plt.plot(range(1,11), wcss, marker = 'o')
plt.title('Elbow Method')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

**Clearly the point to be taken is 3, thus 3 clusters needs to be formed**

# Applying K-Means

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300,
                n_init = 10, random_state = 0 )
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

# Visualising the clusters

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1],
           s = 100, c='yellow', label = 'Iris-Setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1],
           s = 100, c='red', label = 'Iris-Versicolor')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
           s = 100, c='blue', label = 'Iris-Verginica')

#plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[: , 0], kmeans.cluster_centers_[: , 1],
           s = 500, c = 'black', label = 'Centroids', marker = '*')
plt.legend()
```
